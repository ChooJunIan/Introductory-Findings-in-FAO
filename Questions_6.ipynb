```
#9. How should we rank country based on its area harvested, yield and production for all 6 crops(Exploratory)
#- identify top 20% of the countries in each category
#- give a ranking based on their standing
#- calculate the ranking in all 3 attributes for each country
#- provide a final score  for the country

import numpy as np
import pandas as pd

df = pd.read_csv('CLEANED.csv', encoding='latin-1', engine='python')
size_of_row = len(df.index)
size_of_column = len(df.columns)
areaCode = df["Area Code"].value_counts()

```

```
#score based on value count
uniqueAreaCode = df["Area Code"].unique()
areaCode = df["Area Code"].value_counts()/3
areaCodeIndex = areaCode.index

x = areaCode[uniqueAreaCode[0]]

table = pd.DataFrame([x],index = [uniqueAreaCode[0]])
for i in range(len(uniqueAreaCode) - 1):
  x = areaCode[uniqueAreaCode[i+1]]
  new_row = pd.DataFrame([x],index = [uniqueAreaCode[i+1]])
  table = pd.concat([new_row, table]) 

table = table.rename(columns={0: 'Number of estential crops that the country grow'})


```

```
#the graph showing the number of crop that a country grow
import seaborn as sns
import matplotlib.pyplot as plt

sortedTable = table.sort_values(by = ['Number of estential crops that the country grow'])
series = sortedTable.squeeze()
plt.figure(figsize=(50,10))  
ax = sns.barplot(series.index, series).set_title('Number of estentials crop that a country grow')
plt.show()
```

```
#Potatoes: 116
#Lettuce and chicory: 372
#Tomatoes: 388
#Cabbages and other brassicas: 358
#Sugar cane: 156
#Wheat: 15

filter1 = df["Item Code"] == 116
filter2 = df["Item Code"] == 372
filter3 = df["Item Code"] == 388
filter4 = df["Item Code"] == 358
filter5 = df["Item Code"] == 156
filter6 = df["Item Code"] == 15

filter7 = df['Element Code'] == 5312
filter8 = df['Element Code'] == 5419
filter9 = df['Element Code'] == 5510

```

```
#Potatoes Area Harvested

potatoes_df = df[filter1]
potatoes_df_AreaHarvested = potatoes_df[filter7]
size_of_row = len(potatoes_df_AreaHarvested.index)
size_of_column = len(potatoes_df_AreaHarvested.columns)

selectedData = potatoes_df_AreaHarvested.iloc[0:size_of_row,7:size_of_column]
selectedData_size_of_row = len(selectedData.index)
selectedData_size_of_column = len(selectedData.columns)
selectedData_mean = selectedData.mean(axis = 1)
selectedData_mean = selectedData_mean.round(decimals=1)
selectedData_mean = selectedData_mean.sort_values(ascending = True)


selectedData_meanIndex = selectedData_mean.index

for i in range(len(selectedData_meanIndex)):
  temp_filter = potatoes_df_AreaHarvested.index == selectedData_meanIndex[i]
  x = potatoes_df_AreaHarvested[temp_filter].iloc[0][1]


table1 = pd.DataFrame([[x,selectedData_mean[selectedData_meanIndex[0]]]])
for i in range(len(selectedData_meanIndex) - 1):
  temp_filter = potatoes_df_AreaHarvested.index == selectedData_meanIndex[i+1]
  x = potatoes_df_AreaHarvested[temp_filter].iloc[0][1]
  new_row = pd.DataFrame([[x,selectedData_mean[selectedData_meanIndex[i+1]]]])
  table1 = pd.concat([new_row, table1]) 

table1 = table1.rename(columns={0: 'Area Code'})
table1 = table1.rename(columns={1: 'Area Harvested'})
table1.set_index("Area Code", inplace = True)
mean_AreaHarvested = table1
table1 = table1.squeeze()
plt.figure(figsize=(50,10))  
ax = sns.barplot(table1.index, table1).set_title('The mean area harvested for each country from year of 1992 to 2018')
plt.show()
table1
```
