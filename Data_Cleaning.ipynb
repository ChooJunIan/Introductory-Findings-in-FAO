```
#For our report is mainly focus on a country,so we remove all the row that is not representing as a country
#Next since from year 1961 to 1991 consist more than 20% of null data so choose to remove all of the data start from year 1961 to 1992

df5 = pd.read_csv('Production_Crops_E_All_Data_NOFLAG.csv', encoding='latin-1', engine='python')
df5.drop(df5.iloc[:, 7:38], inplace = True, axis = 1)
df5.drop(index=df5.iloc[37790:50251, :].index.tolist(), inplace = True)

df5
```

```
#the data set that removed all those data that is incomplete ,exp doesnt have yield or production

size_of_row = len(df5.index)
size_of_column = len(df5.columns)
elementCode = 1
dropRowIndex = []
for i in range(size_of_row):
  if (elementCode == 1) & (df5.iloc[i][4] == 5312) :
    elementCode = 2
  elif (elementCode == 2) & (df5.iloc[i][4] == 5419) :
    elementCode = 3
  elif (elementCode == 3) & (df5.iloc[i][4] == 5510) :
    elementCode = 1
  else :
    if elementCode == 1 :
      dropRowIndex.append(i)
    elif elementCode == 2 :
      dropRowIndex.append(i)
      dropRowIndex.append(i-1)
    elif elementCode == 3 :
      dropRowIndex.append(i)
      dropRowIndex.append(i-1)
      dropRowIndex.append(i-2)
    elementCode = 1

for i in dropRowIndex:
  df5 = df5.drop(i)

df5

```

```
#for our project we mainly focus on 6 types of crop only,so we try to remove other crop from our dataset
#the crop that we are focusing on our project
#Potatoes: 116
#Lettuce and chicory: 372
#Tomatoes: 388
#Cabbages and other brassicas: 358
#Sugar cane: 156
#Wheat: 15

filter = (df5["Item Code"] == 116) | (df5["Item Code"] == 372) | (df5["Item Code"] == 388) | (df5["Item Code"] == 358) | (df5["Item Code"] == 156) | (df5["Item Code"] == 15)
df5 = df5[filter]
df5
```

```
#remove all those country that grow less than 4 type of the estential crop
dfX = df5["Area Code"].value_counts() 
dfX2 = pd.DataFrame(dfX, columns=['Area Code'])
size_of_rowdfX2 = len(dfX2.index)
x = dfX2.index
y = []
for i in range(size_of_rowdfX2) :
  if dfX2.iloc[i][0] < 12 :
    filter = df5["Area Code"] == x[i]
    temp = df5[filter]
    tempIndex = temp.index
    for i2 in tempIndex :
      y.append(i2)

for i in y:
  df5 = df5.drop(i)

df5


```

```
# remove those row that have more than 60% null data
size_of_row = len(df5.index)
size_of_column = len(df5.columns)

selectedData = df5.iloc[0:size_of_row,7:size_of_column]
selectedData_size_of_row = len(selectedData.index)
selectedData_size_of_column = len(selectedData.columns)
elementCode = 1
test = False 
removingRowIndex = []
selectedDataIndex = selectedData.index
for idx in range(size_of_row):
    if (selectedData.iloc[idx].isnull().sum()) < 11 & (elementCode == 1) & (test == False):
      elementCode = 2
    elif (selectedData.iloc[idx].isnull().sum()) < 11 & (elementCode == 2) & (test == False):
      elementCode = 3
    elif (selectedData.iloc[idx].isnull().sum()) < 11 & (elementCode == 3) & (test == False):
      elementCode = 1
    else :
      if elementCode == 1 :
        removingRowIndex.append(selectedDataIndex[idx])
        elementCode = 2
        test = True
      elif elementCode == 2 :
        if test == True :
          removingRowIndex.append(selectedDataIndex[idx])
        else :
          test = True
          removingRowIndex.append(selectedDataIndex[idx])
          removingRowIndex.append(selectedDataIndex[idx-1])
        elementCode = 3
      elif elementCode == 3 :
        if test == True :
          removingRowIndex.append(selectedDataIndex[idx])
        else :
          removingRowIndex.append(selectedDataIndex[idx])
          removingRowIndex.append(selectedDataIndex[idx-1])
          removingRowIndex.append(selectedDataIndex[idx-2])
        test = False
        elementCode = 1

for i in removingRowIndex:
  df5 = df5.drop(i)
  selectedData = selectedData.drop(i)

df5

```

```
#replace null data to mean value
size_of_row = len(df5.index)
size_of_column = len(df5.columns)

selectedData = df5.iloc[0:size_of_row,7:size_of_column]
selectedData_size_of_row = len(selectedData.index)
selectedData_size_of_column = len(selectedData.columns)
selectedData_mean = selectedData.mean(axis = 1)
selectedData_mean = selectedData_mean.round(decimals=1)

for idx in range(5):
    selectedData.iloc[idx].fillna(value = selectedData_mean.iloc[idx],inplace = True)

df5.iloc[0:size_of_row,7:size_of_column] = selectedData

df5

```
